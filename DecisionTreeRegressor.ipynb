```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.datasets import make_regression
from implementation.DecisionTreeRegression import DecisionTreeRegression
plt.style.use('ggplot')
```

```python
x, y = make_regression(n_features = 1, n_samples = 200, bias = 20, noise = 20)
x = preprocessing.scale(x)
y = preprocessing.scale(y)
plt.scatter(x, y)
```

```python
model = DecisionTreeRegression()
model.fit(x, y, max_depth = 16, threshold_impurity = 0.3)
```

```python

plt.scatter(x, y)
plt.scatter(x, model.predict(x), c = 'b')
plt.title("Decision Tree regression")
plt.show()
```
