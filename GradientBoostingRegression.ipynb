```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.datasets import make_regression
from implementation.GradientBoostingRegression import GradientBoostingRegression
from implementation.DecisionTreeRegression import DecisionTreeRegression
plt.style.use('ggplot')
```

```python
x, y = make_regression(n_features = 1, n_samples = 1000, bias = 20, noise = 20)
x = preprocessing.scale(x)
y = preprocessing.scale(y)
plt.scatter(x, y)

x_test = x[:500]
y_test = y[:500]
x = x[500:]
y = y[500:]
```

```python
model = GradientBoostingRegression()
model.fit(x, y, max_depth = 20, print_every_nth_estimator = 10)
```

```python
model1 = DecisionTreeRegression()
model1.fit(x,y, max_depth = 20)
```

```python

plt.scatter(x_test, y_test, s = 15)
plt.scatter(x_test, model.predict(x_test), c = 'b', s = 15)
plt.scatter(x_test, model1.predict(x_test), c = 'c', s = 15)
plt.legend(["Training Data","Gradient Boosting Predictions","Decision Tree Predictions"])
plt.show()
```
